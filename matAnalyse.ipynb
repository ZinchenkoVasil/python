```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-100,100,200)
plt.plot(x, x**3 - x**2, label='x**3 - x**2')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

исследование функции
1) Область значений: от -бесконечности до +бесконечности
2) Нули ф-ции: x1 = 0; x2 = 1  
3) отрезки знакопостоянства: от -бесконечности до 0 и от 0 до 1 функция принимает отрицательные значения; от 1 до +бесконечности - положительные значения 
4) интервалы монотонности: от -бесконечности до 0 увеличивается; от 0 до 2/3 - уменьшается; от 2/3 до +бесконечности - увеличивается
5) четность: ни четная, ни нечетная функция
6) неограниченна
7) непериодична
