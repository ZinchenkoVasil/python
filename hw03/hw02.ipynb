```python
from scipy.optimize import fsolve
def equations(p):
    x,y = p
    return(y - x**2 + 1, np.exp(x) + x - x*y - 1)
x1, y1 = fsolve(equations,(3,8))
print(x1,y1)
```

```python
x2, y2 = fsolve(equations,(-2,2))
print(x2,y2)
```

значит решение неравенства y = x**2 - 1 и exp(x) + x*(1 - y) > 1
лежит в диапазоне точек:
-1.58183535 < x < 2.6181455 

доказательство, что расстояния при ортогональном преобразовании сохраняются

R1^2 = (x2 - x1)^2 + (y2 - y1)^2 (где R1 - расстояние между 2 точками)

R2 - расстояние между этими точками в другой ортогональной с-ме координат

нам требуется доказать, что R1 = R2

Доказательство:

R2^2 = (a11(x2-x1) + a12(y2 - y1))^2 + (a21(x2-x1) + a22(y2 - y1))^2 

R2^2 = (a12^2 + a22^2)(y2-y1)^2 + (a11^2 + a21^2)(x2-x1)^2 + 2(a11a12 + a21a22)(x2-x1)(y2-y1)

a12^2 + a22^2 = 1, a11^2 + a21^2 = 1, a11a12 + a21a22 = 0

отсуда следует:

R2^2 = (y2-y1)^2 + (x2-x1)^2 = R1^2 (что и требовалось доказать!)
